```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib qt
%matplotlib inline
import missingno as misno
```

```python
df=pd.read_csv("avril line mistake.csv")
df.head(5)
```

Check missing value

```python
df.isnull().sum()
```

fill 0 to NAN values

```python
df=df.fillna(0)
```

```python
df['N° Pcs Checked']=df['N° Pcs Checked'].fillna(0)
df['N° Pcs MISTAKES']=df['N° Pcs MISTAKES'].fillna(0)
df['LINE NUMBER']=df['LINE NUMBER'].fillna('No info')
```

```python
df.isnull().sum()
```

```python
misno.bar(df)
plt.title('Presentation of missing values',fontsize=40,color='red')
```

we have problem of BASIC and Basic the meaning is same but in the system we will have to difrent model we should replace Basic by BASIC

```python
df['Model'].replace(to_replace=['Basic'],value=['BASIC'],inplace=True)
df['LINE NUMBER'].replace(to_replace=['1 BASIC LINE','2 BASIC LINE','3 BASIC LINE','4 BASIC LINE','5 BASIC LINE','11 JOGGER LINE','12 JOGGER LINE','13 JOGGER LINE','15-B JOGGER LINE','16 POLO LINE','17 POLO LINE','18 POLO LINE','6 LEGGING LINE','7 LEGGING LINE','8 TANK TOP LINE','9 TANK TOP LINE','15 JOGGER LINE','14 JOGGER LINE','19 POLO LINE']
                          , value=['line1 BASIC','line2 BASIC','line3 BASIC','line4 BASIC','line5 BASIC','line11 JOGGER','line12 JOGGER','line13 JOGGER','line15-b JOGGER','line16 BASIC','line17 BASIC','line18 BASIC','line6 LEGGING','line7 LEGGING','line8 TANK TOP','line9 TANK TOP','line15 JOGGER','line14 JOGGER','line19 POLO'],inplace=True)
df
```

```python
df.to_csv('forTableau.csv')
```

```python
df.info()
```

```python
df1=df.groupby(['order N°']).aggregate({
    'Model':np.max,
    #'LINE NUMBER':np.max,
    'type of operation':pd.Series.nunique,
    'Element':np.count_nonzero,
    'LINE NUMBER':np.random.choice,
    'N° Pcs Checked':np.sum,
    'N° Pcs MISTAKES':np.sum
    
    })
df1.rename(columns={   'type of operation':'nomber of operation',
                          'Element': 'nomber of operator'},inplace=True)


df1
```

```python
a=df1['N° Pcs MISTAKES'].sum()/df1['N° Pcs Checked'].sum()*100
a
```

We have 6% of mistakes pcs

## Now we will analyze mistak by garment

```python
df3=df.groupby(['Model']).aggregate({
    'N° Pcs Checked'    : np.sum,
    'N° Pcs MISTAKES'   :np.sum
})
df3['percentage']=df3['N° Pcs MISTAKES']/df3['N° Pcs Checked']*100
df3=df3.sort_values('percentage',ascending=False)
df3
```

```python
plt.figure(figsize=(30,30))
plt.xticks(rotation='horizontal')
plt.bar(df3.index,'percentage',data=df3)
```

legging have high rate of mistake comparing by others garment

## We will analyze mistake by each line

```python
df2=df1.groupby(['LINE NUMBER']).aggregate({
    'N° Pcs Checked'    : np.sum,
    'N° Pcs MISTAKES'   :np.sum
})
df2['percentage']=df2['N° Pcs MISTAKES']/df2['N° Pcs Checked']*100
df2=df2.sort_values('percentage',ascending=False)
df2
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation='vertical')
plt.bar(df2.head(10).index,'percentage',data=df2.head(10))
```

we can understand that LINE1 BASIC make more mistake comparing by others

## Analyze mistake by each Operator

we will see the operator how maked more mistakes

```python
df4=df.groupby(['Element']).aggregate({
    'Model':np.max,
    'N° Pcs Checked'    : np.sum,
    'N° Pcs MISTAKES'   :np.sum
})
df4['percentage']=df4['N° Pcs MISTAKES']/df4['N° Pcs Checked']*100
df4=df4[df4['N° Pcs Checked']>=200].sort_values('percentage',ascending=False)
df4.drop(['/'],axis=0,inplace=True)
df5=df4.head(10)
df5
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=df5.index,y='percentage',data=df5,hue='Model',dodge=False)
```

the operator 1003556 have the high percentage of mistakes according to this result we can sad that this element is the bad element this month

and know we will see the best operator this month

```python
#pd.merge(df7,df8,on='NAME OF MISTAKES',how='outer')
df_add1 = df7.add(df8, fill_value=0)
df_add2=df_add1.add(df9, fill_value=0)
df_add3=df_add2.add(df10, fill_value=0)
df_add4=df_add3.add(df11, fill_value=0)
df_add5=df_add4.add(df12, fill_value=0)
df_add5
```
